# Input Parameters

```python
%sql
CREATE WIDGET TEXT catalog_source DEFAULT 'samples';
CREATE WIDGET TEXT catalog_sink DEFAULT 'bronze_dev';

DECLARE OR REPLACE VARIABLE schema_source STRING DEFAULT 'wanderbricks';
```

# Create Schema Sink

```python
%sql
CREATE SCHEMA IF NOT EXISTS IDENTIFIER(:catalog_sink || '.' || schema_source)
```

# Load tables

## Users

```python
%sql
CREATE OR REPLACE TABLE IDENTIFIER(:catalog_sink || '.' || schema_source || '.users')
COMMENT 'Contém dados sobre usuários particulares e empresariais do Wanderbricks.'
TBLPROPERTIES (
  'data.quality.level' = 'Bronze',
  'delta.autoOptimize.optimizeWrite' = 'true',
  'delta.autoOptimize.autoCompact' = 'true'
)
AS
SELECT
  *, 
  convert_timezone('-03:00', current_timestamp()) AS ingestion_timestamp,
  'overwrite' AS ingestion_mode
FROM 
  IDENTIFIER(:catalog_source || '.' || schema_source || '.users');

ALTER TABLE IDENTIFIER(:catalog_sink || '.' || schema_source || '.users') 
ALTER COLUMN ingestion_timestamp COMMENT 'Data e Hora de Ingestão';

ALTER TABLE IDENTIFIER(:catalog_sink || '.' || schema_source || '.users') 
ALTER COLUMN ingestion_mode COMMENT 'Modo de Ingestão'
```
